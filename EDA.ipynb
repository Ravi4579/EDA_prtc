```python
df=df.iloc[:,1:]
```

```python
df
```

```python
df['Temp C'] = pd.to_numeric(df['Temp C'], errors='coerce')
```

```python
#Convert 'Wind' column to int
df['Wind'] = df['Wind'].astype('int64')

# Convert 'Weather' column to categorical
df['Weather'] = df['Weather'].astype('category')
```

```python
df.info()
```

```python
df.shape
```

```python
df[df.duplicated()].shape
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df = df.drop('Temp C', axis=1)
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df = df.rename(columns={'Solar.R': 'Solar'}) # Rename the column
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Ozone'], kde=True)  # Use kde=True for the density curve
plt.title('Distribution of Ozone')
plt.xlabel('Ozone')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Ozone'])
plt.title('Box Plot of Ozone')
plt.ylabel('Ozone')
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Null Values Heatmap')
plt.show()
```

```python
df.columns[df.isnull().any()]
```

```python
df['Ozone'] = df['Ozone'].fillna(df['Ozone'].mean())
```

```python
#fill solar column null values with mean
df['Solar'] = df['Solar'].fillna(df['Solar'].mean())
```

```python
df.columns[df.isnull().any()]
```

```python
# fill weather column null values with mode

# Fill 'Weather' column null values with the mode
df['Weather'] = df['Weather'].fillna(df['Weather'].mode()[0])
```

```python
df.isnull().sum()
```

```python
sns.pairplot(df)
plt.show()
```

```python
df = df.drop('Year', axis=1)
```

```python
# Assuming 'Month' column exists and you want to replace 'May' with 5
df['Month'] = df['Month'].replace('May', 5)
```

```python
df.info()
```

```python
df.iloc[:,0:6].corr()
```

```python
df=pd.get_dummies(df,columns=['Weather'], dtype=int)
df.head()
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming 'df' is your DataFrame and you want to normalize columns 'Ozone', 'Solar', and 'Wind'
columns_to_normalize = ['Ozone', 'Solar', 'Wind']

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit the scaler to the selected columns and transform them
df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

df.head()
```

```python
# Assuming 'df' is your DataFrame and you want to standardize columns 'Ozone', 'Solar', and 'Wind'
from sklearn.preprocessing import StandardScaler

columns_to_standardize = ['Ozone', 'Solar', 'Wind']

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the selected columns and transform them
df[columns_to_standardize] = scaler.fit_transform(df[columns_to_standardize])

df.head()
```
